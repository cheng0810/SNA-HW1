```python
import numpy as np 
import pandas as pd
import networkx as nx
import os
import math
import csv
```

```python
G = nx.Graph()

for node in range(1005):
    G.add_node(node)
```

```python
true_edge = pd.read_csv("/storage/socialnetwork/hw1/training_data.csv")

node1 = true_edge['node1']
node2 = true_edge['node2']


for edge in range(len(node1)):
    G.add_edge(node1[edge],node2[edge])
```

```python
print(G.number_of_nodes(), G.number_of_edges())
```

```python
# preds = nx.jaccard_coefficient(G,  edge)
cn_nb = []
for node in range(len(node1)):
    cn = sorted(nx.common_neighbors(G, node1[node],node2[node]))
    cn_nb.append(len(cn))
```

```python
cn_nb
```

```python
from sklearn import preprocessing 

print(preprocessing.scale(cn_nb))
```

```python
pre_edge = pd.read_csv("/storage/socialnetwork/hw1/predict.csv")

node3 = pre_edge['node1']
node4 = pre_edge['node2']

cn_nb_p = []
for node in range(len(node3)):
    cnp = sorted(nx.common_neighbors(G, node3[node],node4[node]))
    cn_nb_p.append(len(cnp))

ans = preprocessing.scale(cn_nb_p)
```

```python
len(ans)
ans1 = []
```
